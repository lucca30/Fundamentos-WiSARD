```python
from PIL import Image
import numpy as np
im = Image.open('../data/JAFFE/KM.FE1.23.tiff')
pixels = np.array(im)
for i in range(len(pixels)):
    for j in range(len(pixels[1])):
        if(pixels[i][j]>75):
            pixels[i][j] = 1
        else:
            pixels[i][j] = 0
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot
plt.title('Label is {label}'.format(label="teste"))
plt.imshow(pixels, cmap='gray')
plt.show()
```

```python
#com este bloco Ã© possivel importar um txt, e imprimir a imagem que nele possui, apenas para confirmar se houve filtragem
import numpy as np
selected_image = 14
altura = 256
largura = 256
total = altura*largura
path_images = "../data/Jaffetxt_limiar_simples/val_cruz/images_test_2.txt"
path_labels = "../data/Jaffetxt_limiar_simples/val_cruz/labels_test_2.txt"

file_images = open(path_images, 'r')
matrix_images = file_images.readlines()
file_images.close()

file_labels = open(path_labels, 'r')
matrix_labels = file_labels.readlines()
file_labels.close()

image = matrix_images[selected_image]
temp_arr = []
for i in map(str, image[:-1].split(' ')):
    if(i== ''):
        continue
    temp_arr.append(int(i))
temp_arr = np.array(temp_arr).reshape(altura,largura)
import matplotlib.pyplot as plt

# Plot
plt.title('Label is {label}'.format(label=matrix_labels[selected_image]))
plt.imshow(temp_arr, cmap='gray')
plt.show()

    
    
```
